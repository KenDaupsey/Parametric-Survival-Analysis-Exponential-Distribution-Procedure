# *** Exponential Distribution Survival Analysis Procedure ***

```python
### Import Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url= "https://raw.githubusercontent.com/KenDaupsey/Parametric-Survival-Analysis-Exponential-Distribution-Procedure/main/Survival%7EUnemployment%7EDuration.csv"
df= pd.read_csv(url)
df.head()
```

```python
### Exponential Distribution procedure

from lifelines import ExponentialFitter

# Exponential Survival Estimation
expf = ExponentialFitter()
expf.fit(durations=df['spell'], event_observed=df['censor1'])

# Key Metrics for Exponential
print("Median Survival Time (Exponential):", expf.median_survival_time_)
print("Survival Probability at 50 weeks (Exponential):", expf.predict(50))

# Plot Exponential Survival Curve
plt.figure(figsize=(10, 6))
expf.plot_survival_function()
plt.title('Exponential Survival Curve - df_Unemployment2')
plt.xlabel('Time')
plt.ylabel('Survival Probability')
plt.show()
```
